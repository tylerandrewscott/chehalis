```python
import requests
import time
# URL of the website
url = "https://secure.in.gov/apps/idoa/contractsearch/api/contracts/search"
```

```python
start_date = '2000-01-01T00:00:00.0000000'
page_size = 1000
jobj = {#"contractTypeFlags":n,
        "pageNumber":1,"pageSize":page_size,"startDate":start_date}
resp = requests.post(url,json = jobj)
js_result = resp.json()
paginate = js_result['pagination']
total_pages = paginate['totalPages']
result_list = js_result['results']
length = len(result_list)
#if length > 0:
#    print('n =' + str(n))
print(length)
for p in range(2,total_pages):
    print(p)
    jobj = {"pageNumber":p,"pageSize":page_size,"startDate":start_date}
    resp = requests.post(url,json = jobj)
    js_result = resp.json()
    result_list = result_list + js_result['results']
    time.sleep(0.25)
## copy just in case
tlist = result_list
import json
len(tlist)
with open('../../data/raw/indiana_contracts.json','w') as write_file:
    json.dump(tlist,write_file)
        
```
